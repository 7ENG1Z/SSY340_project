```python
import os
import shutil

def delete_half_of_subsubfolders(main_folder):
    # 列出主文件夹下的所有二级子文件夹
    subfolders = [f for f in os.listdir(main_folder) if os.path.isdir(os.path.join(main_folder, f))]

    for subfolder in subfolders:
        subfolder_path = os.path.join(main_folder, subfolder)
        
        # 列出二级子文件夹下的所有三级子文件夹
        subsubfolders = [f for f in os.listdir(subfolder_path) if os.path.isdir(os.path.join(subfolder_path, f))]
        half_length = len(subsubfolders) // 2

        for i in range(half_length):
            subsubfolder_to_delete = os.path.join(subfolder_path, subsubfolders[i])
            shutil.rmtree(subsubfolder_to_delete)
            print(f"Deleted: {subsubfolder_to_delete}")

# 使用方法
main_folder = "/home/lizi/Desktop/MP_Data (copy)"  # 请更改为您的主文件夹路径
delete_half_of_subsubfolders(main_folder)
```

```python

def count_subsubfolders(main_folder):
    # 列出主文件夹下的所有二级子文件夹
    subfolders = [f for f in os.listdir(main_folder) if os.path.isdir(os.path.join(main_folder, f))]

    counts = {}
    for subfolder in subfolders:
        subfolder_path = os.path.join(main_folder, subfolder)
        
        # 列出二级子文件夹下的所有三级子文件夹
        subsubfolders = [f for f in os.listdir(subfolder_path) if os.path.isdir(os.path.join(subfolder_path, f))]
        
        # 将计数保存到字典中
        counts[subfolder] = len(subsubfolders)

    return counts

# 使用方法
main_folder = "/home/lizi/Desktop/MP_Data (copy)"  # 请更改为您的主文件夹路径
folder_counts = count_subsubfolders(main_folder)

for folder, count in folder_counts.items():
    print(f"{folder}: {count} subfolders")

```

```python
import os
import shutil

def move_subsubfolders(src_folder, dest_folder):
    # 列出源文件夹下的所有二级子文件夹
    subfolders = [f for f in os.listdir(src_folder) if os.path.isdir(os.path.join(src_folder, f))]

    for subfolder in subfolders:
        src_subfolder_path = os.path.join(src_folder, subfolder)
        dest_subfolder_path = os.path.join(dest_folder, subfolder)

        # 确保目标二级子文件夹存在
        if os.path.exists(dest_subfolder_path):
            # 列出源二级子文件夹下的所有三级子文件夹
            subsubfolders = [f for f in os.listdir(src_subfolder_path) if os.path.isdir(os.path.join(src_subfolder_path, f))]

            for subsubfolder in subsubfolders:
                src_subsubfolder_path = os.path.join(src_subfolder_path, subsubfolder)
                dest_subsubfolder_path = os.path.join(dest_subfolder_path, subsubfolder)

                # 检查目标位置是否存在同名的三级子文件夹
                counter = 1
                original_name = subsubfolder
                while os.path.exists(dest_subsubfolder_path):
                    subsubfolder = f"{original_name}_{counter}"
                    dest_subsubfolder_path = os.path.join(dest_subfolder_path, subsubfolder)
                    counter += 1

                # 移动三级子文件夹
                shutil.move(src_subsubfolder_path, dest_subsubfolder_path)
                print(f"Moved {src_subsubfolder_path} to {dest_subsubfolder_path}")

# 使用方法
src_folder = "/home/lizi/Desktop/MP_Data (another copy)"  # 请更改为 folderA 的路径
dest_folder = "/home/lizi/Desktop/WLASL_agument (final) (another copy)"  # 请更改为 folderB 的路径
move_subsubfolders(src_folder, dest_folder)
```

```python
import os

def rename_third_level_folders_to_incremental_numbers(main_folder):
    # 获取所有二级子文件夹
    subfolders = [f for f in os.listdir(main_folder) if os.path.isdir(os.path.join(main_folder, f))]

    for subfolder in subfolders:
        subfolder_path = os.path.join(main_folder, subfolder)
        
        # 获取二级子文件夹中的所有三级子文件夹
        third_level_folders = [f for f in os.listdir(subfolder_path) if os.path.isdir(os.path.join(subfolder_path, f))]

        counter = 0
        for third_level_folder in third_level_folders:
            third_level_folder_path = os.path.join(subfolder_path, third_level_folder)
            new_name = str(counter)
            new_third_level_folder_path = os.path.join(subfolder_path, new_name)

            # 确保新名称不会与已存在的文件夹冲突
            while os.path.exists(new_third_level_folder_path):
                counter += 1
                new_name = str(counter)
                new_third_level_folder_path = os.path.join(subfolder_path, new_name)

            # 重命名三级文件夹
            os.rename(third_level_folder_path, new_third_level_folder_path)
            print(f"Renamed {third_level_folder_path} to {new_third_level_folder_path}")
            counter += 1

# 使用方法
main_folder = "/home/lizi/Desktop/WLASL_agument(combine)"  # 请更改为包含二级文件夹的主文件夹路径
rename_third_level_folders_to_incremental_numbers(main_folder)
```
